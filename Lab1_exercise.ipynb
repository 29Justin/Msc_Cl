## Exercise 6
- Sort the list alphabetically.
- Count how many words are present.
- Loop through the list and print each word.
- Find the longest word in ["apple","banana","computer","AI","language"]

```python
words = ["apple", "banana", "computer", "AI", "language"]

# 1. Sort the list alphabetically
words_sorted = sorted(words, key=lambda x: x.lower())
print("Sorted list:", words_sorted)

# Count how many words are present


# 2. Count how many words are present
word_count = len(words[0])
print("Total words:", word_count)
word_count = len(words[1])
print("Total words:", word_count)
word_count = len(words[2])
print("Total words:", word_count)
word_count = len(words[3])
print("Total words:", word_count)

word_count = len(words[4])
print("Total words:", word_count)


# 3. Loop through the list and print each word
print("\nWords in list:")
for word in words:
    print(word)

# 4. Find the longest word
longest_word = max(words, key=len)
print("\nLongest word:", longest_word)
```

## Final Exercise

- Read the file and print the total number of words.
- Build a dictionary mapping each word to its length (ignore punctuation-only tokens).
- Print the 5 longest words in the file.
- Print how many words are longer than 6 characters.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import string

# 1. Read the file
with open("/sample.txt", "r") as f:
    text = f.read()

# Clean words: split by space, remove punctuation, and ignore empty tokens
words = []
for word in text.split():
    clean_word = word.strip(string.punctuation).lower()
    if clean_word:  # keep if it's not empty
        words.append(clean_word)

# 2. Print total number of words
print("Total words:", len(words))

# 3. Build word-to-length dictionary
word_lengths = {word: len(word) for word in words}
print("Word lengths dictionary:", word_lengths)

# 4. Print 5 longest unique words
longest_5 = sorted(word_lengths.keys(), key=len, reverse=True)[:5]
print("5 longest words:", longest_5)

# 5. Print count of words longer than 6 characters
long_words = [w for w in words if len(w) > 6]
print("Words longer than 6 chars:", len(long_words))
```
